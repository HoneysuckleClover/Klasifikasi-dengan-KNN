```python
import pandas as pd #memanggil file dari luar
import numpy as np #komputasi
import matplotlib.pyplot as plt #buat diagram
```

```python
data = pd.read_csv('/content/Social_Network_Ads.csv')
x = data.iloc[:, [3,4]].values #untuk memanggil kolom 2 dan 3
y = data.iloc[:, -1].values #untuk memanggil semua kolom y
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier #pakai algoritma K-NN
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix #untuk mengetahui akurasinya
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1,x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                    np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('RED', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
              c = ListedColormap(('red', 'GREEN'))(i), label = j)
plt.title('Klasifikasi Data dengan K-NN (Data Training)')
plt.xlabel('Estimasi Gaji')
plt.ylabel('Purchased')
plt.legend()
plt.show()
```
